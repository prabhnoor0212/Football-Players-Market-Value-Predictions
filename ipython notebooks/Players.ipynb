```python
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
records=[]
for index in range(1,21):
    url='https://www.transfermarkt.com/laliga/marktwertaenderungen/wettbewerb/ES1/pos//detailpos/0/verein_id/0/land_id/0/plus/1/page/%s'%str(index)
    res=requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text,'html.parser')
    r4 = soup.find_all('tr',attrs={'class':'odd'})
    r3 = soup.find_all('tr',attrs={'class':'even'})
    for r in r4:
        name = r.find('img')['title']
        position = r.select('td')[3].text
        nationality = r.select('td')[4].find('img')['title']
        age = int(r.select('td')[5].text)
        club = r.select('td')[6].find('img')['alt']
        curr_val = r.select('td')[7].text[:-3]
        prev_val = r.select('td')[7].find('span')['title'].split(': ')[1].split(';')[0]
        records.append((name,position,nationality,age,club,curr_val,prev_val))
    for r in r3:
        name = r.find('img')['title']
        position = r.select('td')[3].text
        nationality = r.select('td')[4].find('img')['title']
        age = int(r.select('td')[5].text)
        club = r.select('td')[6].find('img')['alt']
        curr_val = r.select('td')[7].text[:-3]
        prev_val = r.select('td')[7].find('span')['title'].split(': ')[1].split(';')[0]
        records.append((name,position,nationality,age,club,curr_val,prev_val))
```
