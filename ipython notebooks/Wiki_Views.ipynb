```python

def impute_Wiki(cols):
    url = cols[0]
    browser = webdriver.Chrome()  
    browser.get(url)  
    time.sleep(7)
    html_source = browser.page_source  
    browser.quit()

    soup = BeautifulSoup(html_source,'html.parser') 
    try:
        comments = soup.find_all('span',{'class':'pull-right'})  
        res = (comments[5].text.split('\n')[1].replace(' ',''))
    except IndexError:
        res = '0'
    results.append(res)
    print(res+" "+str(len(results)))
```
