```python
records = []
for index in range(1,13): 
    url = 'https://www.transfermarkt.co.uk/laliga/torschuetzenliste/wettbewerb/ES1/ajax/yw1/saison_id/2016/altersklasse/alle/detailpos/0/plus/1/page/%s'%str(index)
    r = requests.get(url,headers=headers)
    soup = BeautifulSoup(r.text,'html.parser')
    r1 = soup.find_all('tr',attrs={'class':'odd'})
    r2 = soup.find_all('tr',attrs={'class':'even'})
    for r in r1:
        name = r.select('img')[0]['title']
        club_apps = int(r.select('td')[8].text)
        assists = int(r.select('td')[9].text)
        mins_club = int(r.select('td')[11].text.replace('.','').replace("'",""))
        mins_per_goal = float(r.select('td')[12].text.replace("'",""))
        goal_per_game= float(r.select('td')[13].text.replace(',','.'))
        goals = int(r.select('td')[14].text)
        records.append((name,club_apps,assists,mins_club,mins_per_goal,goal_per_game,goals))
        
    for r in r2:
        name = r.select('img')[0]['title']
        club_apps = int(r.select('td')[8].text)
        assists = int(r.select('td')[9].text)
        mins_club = int(r.select('td')[11].text.replace('.','').replace("'",""))
        mins_per_goal = float(r.select('td')[12].text.replace("'",""))
        goal_per_game= float(r.select('td')[13].text.replace(',','.'))
        goals = int(r.select('td')[14].text)
        records.append((name,club_apps,assists,mins_club,mins_per_goal,goal_per_game,goals))  
```

```python
def imputeRat(cols):
    name = cols[0]
    url = 'http://www.futhead.com/18/players/?name=%s&sort=vote_sum&bin_platform=ps'%name  
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'html.parser')
    res = soup.find_all('div',attrs={'class':'content'})
    if len(res) != 1:
        fir = res[1]
        f_rating = int(fir.select('span')[1].text)
        f_pace = int(fir.select('span')[7].text)
        f_shooting = int(fir.select('span')[10].text)
        f_passing = int(fir.select('span')[13].text)
        f_drib = int(fir.select('span')[16].text)
        f_def = int(fir.select('span')[19].text)
        f_phy = int(fir.select('span')[22].text)
        f_wrate = fir.select('span')[34].text
        return f_rating,f_pace,f_shooting,f_passing,f_drib,f_def,f_phy,f_wrate
    else:
        return np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan   
```

```python
recgoal['F_Rating'],recgoal['F_Pace'],recgoal['F_Shooting'],recgoal['F_Passing'],recgoal['F_Drib'],recgoal['F_Def'],recgoal['F_Phy'],recgoal['F_WRate'] = recgoal[['Name']].head().apply(imputeRat,axis=1)
```

```python
def impt(cols):
    name = cols[0]
    url = 'http://www.futhead.com/18/players/?name=%s&sort=vote_sum&bin_platform=ps'%name
    return url
```

```python
recgoal['URL'] = recgoal[['Name']].apply(impt,axis=1)
```

```python
rec=[]
def imputeRat(cols):
    url = cols[0]
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'html.parser')
    res = soup.find_all('div',attrs={'class':'content'})
    if len(res) != 1:
        fir = res[1]
        f_rating = int(fir.select('span')[1].text)
        f_pace = int(fir.select('span')[7].text)
        f_shooting = int(fir.select('span')[10].text)
        f_passing = int(fir.select('span')[13].text)
        f_drib = int(fir.select('span')[16].text)
        f_def = int(fir.select('span')[19].text)
        f_phy = int(fir.select('span')[22].text)
        f_wrate = fir.select('span')[34].text
        rec.append([f_rating,f_pace,f_shooting,f_passing,f_drib,f_def,f_phy,f_wrate])
    else:
        rec.append([np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan ])   
```

```python
recgoal[['URL']].apply(imputeRat,axis=1)
```

```python
fifa = pd.DataFrame(rec,columns=['F_Rating','F_Pace','F_Shooting','F_Passing','F_Drib','F_Def','F_Phy','F_WRate'])
```
